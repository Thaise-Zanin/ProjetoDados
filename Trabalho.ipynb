# Etapa 2 – Concatenação dos Dados de Criminalidade (Balão 2)

```python
import pandas as pd

caminho_arquivo = "/content/drive/MyDrive/ProjetoDados/Dados/Crimes/2024.csv"

# criando uma função pra padronizar as colunas e depois aplicar em tds os csvs
def tratar_df(caminho_arquivo):

    df = pd.read_csv(caminho_arquivo, sep=';', encoding='latin1')

    df.columns = df.columns.str.lower().str.replace(' ', '_')  # colocando o nome das colunas em letra minúscula e substituindo o espaço do meio pelo underline
    df = df.loc[:, ~df.columns.str.contains('unnamed')]  # pegando só as colunas q não sao unnamed
    df = df.drop(columns=['...'], errors='ignore')  # removendo essa coluna dos tres pontinhos

    df['data_fato'] = pd.to_datetime(df['data_fato'], errors='coerce')  # mudando o tipo da coluna de data

    # mudando o tipo das colunas pra economizar memória
    colunas_categoria = [
        'grupo_fato',
        'tipo_enquadramento',
        'tipo_fato',
        'municipio_fato',
        'local_fato',
        'sexo_vítima',
        'cor_vítima'
    ]

    for col in colunas_categoria:
        df[col] = df[col].astype('category')

    # mudando o tipo das colunas de quantidade e idade pra inteiro
    df['quantidade_vítimas'] = df['quantidade_vítimas'].astype('Int64')
    df['idade_vítima'] = df['idade_vítima'].astype('Int64')

    return df
```

```python
df_2021 = tratar_df("/content/drive/MyDrive/ProjetoDados/Dados/Crimes/2021.csv")
df_2022 = tratar_df("/content/drive/MyDrive/ProjetoDados/Dados/Crimes/2022.csv")
df_2023 = tratar_df("/content/drive/MyDrive/ProjetoDados/Dados/Crimes/2023.csv")
df_2024 = tratar_df("/content/drive/MyDrive/ProjetoDados/Dados/Crimes/2024.csv")
df_2025 = tratar_df("/content/drive/MyDrive/ProjetoDados/Dados/Crimes/2025.csv")
df_2026 = tratar_df("/content/drive/MyDrive/ProjetoDados/Dados/Crimes/2026.csv")
```

```python
# pra verificar se os tipos de dados ficaram certinhos

df_2021.info()
```

```python
# concatenando (juntando) tds os dfs de crimes

df_crimes = pd.concat([
    df_2021,
    df_2022,
    df_2023,
    df_2024,
    df_2025,
    df_2026], ignore_index=True)
```

```python
# testando se foi td concatenado e os tipos das colunas estão certos

df_crimes.info()
df_crimes
```

```python
df_pf = df_crimes[df_crimes["municipio_fato"] == "PASSO FUNDO"] #filtrando apenas os crimes do município de Passo Fundo
df_pf
```

Etapa 3 – Limpeza e Padronização (Balão 3)

```python
# verificando se tem coisa duplicada (já verifiquei e não tem, não precisar usar o drop!!!!)

df_pf.duplicated().sum()
```

```python
# verificando se tem dados vazios

df_pf.isna().sum()
```

```python
# apagando valores nulos quando são poucos faltantes

df_pf = df_pf.dropna(subset=[
    'tipo_fato',
    'municipio_fato',
    'local_fato'
])
```

```python
df_pf = df_pf.copy()

# alterando os valores nulos de bairro para "desconhecido"
df_pf.loc[:, 'bairro'] = df_pf['bairro'].fillna('desconhecido')

# alterando os valores nulos de sexo_vítima para "desconhecido"
if 'desconhecido' not in df_pf['sexo_vítima'].cat.categories:
    df_pf.loc[:, 'sexo_vítima'] = df_pf['sexo_vítima'].cat.add_categories('desconhecido')

df_pf.loc[:, 'sexo_vítima'] = df_pf['sexo_vítima'].fillna('desconhecido')

# alterando os valores nulos de cor_vítima para "desconhecido"
if 'desconhecido' not in df_pf['cor_vítima'].cat.categories:
    df_pf.loc[:, 'cor_vítima'] = df_pf['cor_vítima'].cat.add_categories('desconhecido')

df_pf.loc[:, 'cor_vítima'] = df_pf['cor_vítima'].fillna('desconhecido')

# idade (média inteira)
media_idade = round(df_pf["idade_vítima"].mean())

df_pf.loc[:, "idade_vítima"] = df_pf["idade_vítima"].fillna(media_idade).astype("Int64")
```

```python
# verificando o tratamento dos dados vazios

df_pf.isna().sum()
```

```python
df_pf['tipo_enquadramento'].unique()
```

```python
# corrigindo inconsistências nas colunas de forma geral

import unicodedata

# função pra transformar tudo em minúsculo, tirar espaços extras e remover acentos

def normalizar(texto):
    if pd.isna(texto):
        return texto
    texto = texto.lower().strip()
    return unicodedata.normalize('NFKD', texto).encode('ascii', 'ignore').decode('utf-8')

# aplicando a função nas colunas
colunas_categoria = [
    'grupo_fato',
    'tipo_enquadramento',
    'tipo_fato',
    'municipio_fato',
    'local_fato',
    'sexo_vítima',
    'cor_vítima'
]

for col in colunas_categoria:
    df_pf[col] = df_pf[col].apply(normalizar)

# apagando as linhas onde não constam a data do crime

df_pf = df_pf.dropna(subset=['data_fato'])
```

```python
df_pf.isna().sum()
```

```python
crimes_homicidio = [
    'homicidio doloso',
    'homicidio culposo',
    'homicidio culposo direcao veic automotor',
    'feminicidio',
    'feminicidio art 121 par 2 vi'
]

df_pf_homicidios = df_pf[df_pf['tipo_enquadramento'].isin(crimes_homicidio)]
```

# Etapa 4 – Integração com Dados Meteorológicos (Balão 4)


```python
# importar o dataset meteorológico
df_clima = pd.read_csv(
    "/content/drive/MyDrive/ProjetoDados/Dados/Clima/passo_fundo_2021-01-01_2025-09-05.csv",
    encoding='latin1',
    sep=';',
    skiprows=10
)
```

```python
# visualização do dataset de clima
df_clima.head()
df_clima.info()
df_clima.columns
```

```python
# remove coluna 6 unnamed
df_clima.drop(columns=['Unnamed: 6'], inplace=True)
```

```python
# renomeia colunas
df_clima.columns = [
    'data',
    'precipitacao',
    'temp_max',
    'temp_min',
    'umidade',
    'vento'
]
```

```python
# ajusta data que estava como Dtype object
df_clima['data'] = pd.to_datetime(
    df_clima['data'],
    dayfirst=True,
    format='mixed',
    errors='coerce'
)
df_clima.info()
```

```python
# converte essas colunas para numero
cols = ['precipitacao', 'temp_max', 'temp_min', 'umidade', 'vento']

for col in cols:
    df_clima[col] = pd.to_numeric(df_clima[col], errors='coerce')
df_clima.info()
```

```python
# valores ausentes foram preenchidos, visto que o clima de ontem ou de amanhã é parecido
df_clima.fillna(method='ffill', inplace=True)
df_clima.fillna(method='bfill', inplace=True)
df_clima.isnull().sum()
```

```python
# Padronizar a data
df_pf_homicidios['data'] = pd.to_datetime(df_pf_homicidios['data_fato']).dt.normalize()
df_clima['data'] = pd.to_datetime(df_clima['data']).dt.normalize()
```

```python
# Cria coluna cidade igual para os dois

df_pf_homicidios['cidade'] = 'passo fundo'
df_clima['cidade'] = 'passo fundo'
```

```python
# faz o merge entre clima e crimes
df_final = pd.merge(
    df_pf_homicidios,
    df_clima,
    on=['data', 'cidade'],
    how='left'
)
df_final.info()
```

```python
df_final.head()
```

```python
# irá remover as linhas que não tiveram clima
df_final = df_final.dropna(subset=['precipitacao', 'temp_max', 'temp_min', 'umidade', 'vento'])
```

```python
# verifica se nenhuma coluna gerou dados vazios
df_final.isnull().sum()
```

```python
# verificando se as colunas de data e cidade possuem as mesmas infos de data_fato e municipio_fato, pra depois excluir

(df_final['municipio_fato'] == df_final['cidade']).all()
(df_final['data_fato'] == df_final['data']).all()

```

```python
# como as informações das colunas são iguais, vou apagar data e cidade pra evitar duplicidade
# para deixar o df mais limpo, vou excluir a coluna "sequência", "local_fato" e "bairro", por não possuírem relevância para a nossa análise
df_final = df_final.drop(columns=['data', 'cidade', 'sequência', 'local_fato', 'bairro'])
```

```python
# a fim de melhor visualização, vou criar a coluna "chuva" pra verificar se choveu ou não no dia em que o crime ocorreu. Acho que apenas "precipitacao" em número não deixa muito claro.
# se a precipitação for maior que zero, significa que houve chuva e na coluna deve constar "sim"

df_final['chuva'] = (df_final['precipitacao'] > 0).map({True: 'Sim', False: 'Não'})
```

```python
df_final
```

Foi realizada a integração entre os dados de criminalidade e os dados meteorológicos utilizando as colunas data e cidade como chave de junção, considerando o município de Passo Fundo.

Inicialmente, o dataset meteorológico apresentava valores ausentes nas variáveis climáticas (precipitação, temperatura, umidade e vento). Esses valores foram tratados utilizando as técnicas de forward fill e backward fill, assumindo que as condições climáticas variam de forma gradual entre dias consecutivos.

Após o merge, alguns registros permaneceram com valores ausentes nas variáveis climáticas devido à inexistência de dados meteorológicos para determinadas datas no período analisado.

Os dados de criminalidade não foram alterados, pois representam eventos reais e não devem ser estimados ou preenchidos artificialmente.

Dessa forma, optou-se por remover apenas os registros que não possuíam informações meteorológicas correspondentes, garantindo a consistência e a confiabilidade do dataset final.

Assim, obteve-se um dataset unificado contendo variáveis de criminalidade e clima, sem valores nulos nas colunas meteorológicas.

# Etapa 5 – Transformações (Balão 5)

```python
# Como não vou trabalhar com muitas das colunas categóricas, para melhor visualização do dataset vou optar por deixá-las como categoria mesmo

# A coluna de chuva também já foi criada com o intuito de facilitar a visualização da precipitação, que antes era apenas numérica e pouco legível

# Para reduzir ruído, vou unificar em "tipo_enquadramento" o 'feminicidio' e 'feminicidio art 121 par 2 vi' pois são a mesma coisa

# Assim, localizo todas as linhas que contém a palavra feminicidio e substituo o valor inteiro da célula por apenas "feminicidio"

df_final.loc[
    df_final['tipo_enquadramento'].str.contains('feminicidio'),
    'tipo_enquadramento'
] = 'feminicidio'
```

```python
# Quanto aos outliers, vou verificar referente as idades das vítimas, pelo IQR

Q1 = df_final['idade_vítima'].quantile(0.25)
Q3 = df_final['idade_vítima'].quantile(0.75)
IQR = Q3 - Q1

limite_superior = Q3 + 1.5 * IQR
limite_inferior = Q1 - 1.5 * IQR

print(Q1, Q3, IQR)
print(limite_superior, limite_inferior)

# pra visualizar os outliers

outliers = df_final[
    (df_final['idade_vítima'] < limite_inferior) |
    (df_final['idade_vítima'] > limite_superior)
]

outliers

# Como não obtive nenhum resultado fora dos limites, verifiquei que não existem outliers na idade das vítimas
```

```python
# para fins de padronização, vou remover os acentos nos nomes das colunas novamente, usando a função que já havia criado na etapa 3

df_final.columns = [normalizar(col) for col in df_final.columns]
df_final

# Quanto à normalização, optei por não fazer pq meus dados perderiam o sentido e dificultaria muito a interpretação. Como o objetivo do trabalho é fazer uma análise, não cabe aqui normalizar informações como idade, temperatura e etc
```

```python
# Salvando o df final em um arquivo csv para entrega

df_final.to_csv('df_final.csv', index=False, encoding='utf-8')
```

Foram realizadas padronizações no dataset para melhorar a qualidade e facilitar a análise. As categorias da coluna tipo_enquadramento foram unificadas (como os casos de feminicídio), reduzindo inconsistências. Também foi feita a remoção de acentos nos nomes das colunas para manter um padrão.

Em relação aos outliers, foi aplicada a técnica de IQR na variável idade das vítimas, e não foram identificados valores fora dos limites, indicando ausência de outliers.

A normalização dos dados não foi realizada, pois poderia prejudicar a interpretação dos valores no contexto da análise.
